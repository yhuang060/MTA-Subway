```python
import pandas as pd
```

```python
df = pd.concat(map(pd.read_csv, ['http://web.mta.info/developers/data/nyct/turnstile/turnstile_210102.txt']))
df.shape
```

```python
#Create a column 'Hour'
df['TIME'].unique()
df['Hour']= list(map(lambda x: x.split(':')[0],df['TIME'][:]))
df.head()

#rename column name of 'EXITS'
df.columns=['C/A', 'UNIT', 'SCP', 'STATION', 'LINENAME', 'DIVISION', 'DATE', 'TIME',
       'DESC', 'ENTRIES','EXITS','Hour']
df.head()
df[df['STATION']=='1 AV'].head(30)
```

```python
#change Date format
from datetime import datetime
df['DATE']=list(map(lambda df: datetime.strptime(df,'%m/%d/%Y'),df['DATE']))
```

```python
print(df['C/A'].unique().shape)
print(df['UNIT'].unique().shape)
print(df['SCP'].unique().shape)
print(df['STATION'].unique().shape)
print(df['LINENAME'].unique().shape)
```

```python


#df[df['STATION']=='1 AV'].sort_values(['DATE']).SCP.unique()
#len(df[df['STATION']=='1 AV'].sort_values(['DATE']))
```

```python
#df1=pd.DataFrame(df.groupby(['STATION','DATE']).agg({'ENTRIES':'max','EXITS':'max'}).reset_index().sort_values(['STATION','DATE']))
df1=pd.DataFrame(df.groupby(['STATION','SCP','DATE']).agg({'ENTRIES':'max','EXITS':'max'})
                 .reset_index().sort_values(['STATION','DATE']))

#add columne indicate each date's entries volume and Exit volume
df1["ENTRIES_date"]=0
for i in range(1,len(df1)):
    if df1.loc[i,'STATION']==df1.loc[(i-1),'STATION'] and df1.loc[i,'SCP']==df1.loc[(i-1),'SCP']:
        df1.loc[i,"ENTRIES_date"]=df1.loc[i,'ENTRIES']-df1.loc[(i-1),'ENTRIES']
        

df1["EXITS_date"]=0
for i in range(1,len(df1)):
    if df1.loc[i,'STATION']==df1.loc[(i-1),'STATION'] and df1.loc[i,'SCP']==df1.loc[(i-1),'SCP']:
        df1.loc[i,"EXITS_date"]=df1.loc[i,'EXITS']-df1.loc[(i-1),'EXITS']

#add column 'Daily_Volume'for total daily Volume (ENTRIES_date+EXITS_date)
df1['Daily_Volume']=df1["ENTRIES_date"]+df1["EXITS_date"]

df1.head(14)

#Then we remove the row with value 0 (no data for the date before- not able to calculate)
df1=df1[df1['ENTRIES_date']>0]

df1=df1.drop(columns=['ENTRIES','EXITS'])
df1.head()
```

```python
#Merge df1 & df by STATION & DATE column

update_df=pd.merge(df,df1,on=["STATION",'SCP',"DATE"])
update_df[update_df['STATION']=='1 AV'].head(30)

#update_df.shape #(153907, 15)
```

```python
#add column for weekday
#from datetime import datetime
update_df["DATE"]= update_df["DATE"].weekday()
```
